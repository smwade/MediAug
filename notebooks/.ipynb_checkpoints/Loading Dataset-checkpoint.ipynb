# Load Dataset

A `Dataset` in MediAug is a class to manage image and mask datasets. There are several usefull operations like seeing the classes and size. Most importantly this can be used in `Pipelines` for creating synthetic datasets.

````python
from mediaug.dataset import Dataset
from mediaug.visulize import plot_img_and_mask 
````

## Load the Dataset

Internally a `Dataset` is just a directory with a special structure. This
is

```
cells/
  image/
    class1/
      img1.png
      img2.png
      ...
     class1/
      img1.png
      img2.png
      ...
  mask/
    class1/
      img1.png
      img2.png
      ...
     class1/
      img1.png
      img2.png
      ...
```

````python
cells = Dataset('/Users/sean/projects/cancerDetection/data/cells')
````

## Explore the Data

````python
print('Number of cells:', cells.size)
````

````python
print('Cell Types:', cells.classes)
````

## Get a Random Sample from the Dataset

````python
x = cells.random_sample()
````

````python
plot_img_and_mask(x.img, x.mask)
````
